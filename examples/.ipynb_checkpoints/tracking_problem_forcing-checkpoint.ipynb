```python
class EinsumLayer(tf.keras.layers.Layer):
    """
    Layer wrapping a single tf.einsum operation.

    Usage:
    psi_y = EinsumLayer("ij,ijk->ik")((psi_x, K_u))
    """

    def __init__(self, equation: str):
        super().__init__()
        self.equation = equation

    def call(self, psi_x, K_u):
        return tf.einsum(self.equation, psi_x, K_u)

    def get_config(self):
        return {"equation": self.equation}
```

```python
def generate_model():
    layer_sizes=[128, 256, 128]
    hidden_layers = [Dense(s, activation='tanh') for s in layer_sizes]
    # hidden_layers = [Dense(s) for s in layer_sizes]
    output_layer = Dense(n_phi)
    
    inputs_u = Input((param_dim,))
    hidden_u = inputs_u
    for layer in hidden_layers:
        hidden_u = layer(hidden_u)   
    phi_u = output_layer(hidden_u)
    K_u_entry = Dense(n_psi**2, use_bias=True)(phi_u)
    K_u = tf.reshape(K_u_entry, shape=(-1, n_psi, n_psi))
    model_K_u = Model(inputs=inputs_u, outputs=K_u, name='K_u')
    
    inputs_u = Input((param_dim,))
    inputs_psi_x = Input((n_psi,))
    K_u = model_K_u(inputs_u)
    psi_y = EinsumLayer('ij,ijk->ik')(inputs_psi_x, K_u)
    model_K_u_pred = Model(inputs=[inputs_u, inputs_psi_x], outputs=psi_y, name='K_u_pred')
    
    dic_func = PsiNN(layer_sizes=[100,100,100], n_psi_train=n_psi_train)
    
    # Build the model related to Psi
    inputs_x = Input((target_dim,))
    inputs_y = Input((target_dim,))
    inputs_u = Input((param_dim,))

    psi_x = dic_func(inputs_x)
    psi_y = dic_func(inputs_y)

    psi_next = model_K_u_pred([inputs_u, psi_x])

    outputs = psi_next - psi_y
    model = Model(inputs=[inputs_x, inputs_y, inputs_u], outputs=outputs)
    return model, model_K_u, model_K_u_pred, dic_func
```

```python
model_pk, model_K_u_pk, model_K_u_pred_pk, dic_func_pk = generate_model()
```

```python
zeros_data_y_train = tf.zeros_like(dic_func_pk(data_y))
zeros_data_y_valid = tf.zeros_like(dic_func_pk(data_y_valid))
```

```python
def generate_prbs(traj_len):
    signal = np.zeros(traj_len)
    y = np.random.uniform(low=0, high=1, size=(traj_len,))

    for i in range(y.shape[0]):
        if y[i] > 1 - 0.5:
            signal[i] = 1
    prbs = 2 * signal -1
    return prbs
```

```python
traj_len_test = 305
prbs = generate_prbs(traj_len = traj_len_test)
prbs_control = tf.reshape(prbs, shape=(-1,u_dim))
```

```python
def generate_control_soln(x_init, traj_len, control):
    data_true = [x_init]
    
    for i in range(traj_len):
        data_next = bilinear_motor.generate_next_data(data_true[-1], control[i])
        data_true.append(data_next)

    data_true = tf.squeeze(tf.convert_to_tensor(data_true))
    return data_true
```

```python
def compute_koopman_pred(x_init, model_K_u_pred, control, dic_func, traj_len):
    data_pred = [x_init]
    psi_x_init = dic_func(x_init)
    psi_x_list = [psi_x_init]

    B = dic_func.generate_B(data_x)

    for i in range(traj_len):
        psi_y = model_K_u_pred([control[i], psi_x_list[-1]])
        y_pred = psi_y @ B
        data_pred.append(y_pred)
        psi_x_list.append(psi_y)

    data_pred = tf.squeeze(tf.convert_to_tensor(data_pred))
#     psi_x_list = tf.squeeze(tf.convert_to_tensor(psi_x_list))
    return data_pred
```

# Inverse Problem

```python
traj_len_test_control = 305
```

```python
B = dic_func_pk.generate_B(data_x)
B = tf.reshape(B[:,1], (-1,1))
```

```python
def mpc_loss(param, tau, model_K_u_pred, x_true_list, x_pred_list, dic_func, current_time):
    param = tf.reshape(param, shape=(param.shape[0], 1, u_dim))
    loss_list = []
    psi_next_list = [dic_func(tf.reshape(x_pred_list[-1], shape=(-1,x_dim)))]
    x_true_list = tf.reshape(x_true_list, shape=(-1, 1))

    for i in range(tau):
        psi_current = psi_next_list[-1]
        K_psi_next = model_K_u_pred([param[i], psi_current])
        psi_next_list.append(K_psi_next)
        # Use x2 to build the loss function       
        loss_current = tf.norm(K_psi_next @ B - x_true_list[current_time+i+1])
        loss_list.append(loss_current)
    loss= tf.reduce_sum(loss_list)
    return loss

def generate_cons_lhs(i):
    def cons_func_lhs(u):
        return u[i] + 1
    return cons_func_lhs

def generate_cons_rhs(i):
    def cons_func_rhs(u):
        return - u[i] + 1
    return cons_func_rhs

```

# Case 2: Time-varying Reference

## Add constraints for $y$

```python
obs_matrix = np.zeros(shape=(2,1))
obs_matrix[1,:] = 1
obs_matrix
```

```python
def generate_cons_y_lhs(x):
    def cons_y(u):
        u = tf.reshape(u, shape=(u.shape[0], u_dim))
        x_next = bilinear_motor.generate_next_data(tf.reshape(x, shape=(-1,x_dim)), u)
        y = x_next @ obs_matrix
        return np.squeeze(y) + 0.4
    return cons_y

def generate_cons_y_rhs(x):
    def cons_y(u):
        u = tf.reshape(u, shape=(u.shape[0], u_dim))
        x_next = bilinear_motor.generate_next_data(tf.reshape(x, shape=(-1,x_dim)), u)
        y = x_next @ obs_matrix
        return -np.squeeze(y) + 0.4
    return cons_y
```

```python
def generate_cos_ref_soln(traj_len):
    t = np.linspace(0, 1, traj_len, endpoint=False)
    control = 0.5 * np.cos(2 * np.pi * t ) # Here I do not /3 because have scaled in t (np.linspace)
    return control
```
