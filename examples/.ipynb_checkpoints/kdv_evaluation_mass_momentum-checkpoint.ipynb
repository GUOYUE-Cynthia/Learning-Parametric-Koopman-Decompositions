```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot setting
linear_curve = 'darkorange'
linear_shadow = 'navajowhite'
bilinear_curve = 'forestgreen'
bilinear_shadow = 'darkseagreen'
pk_curve = 'darkslateblue'
pk_shadow = 'lightsteelblue'
ref_color = 'black'


exact_color = 'darkred'
linear_no_penalty_color = 'gold'
bilinear_no_penalty_color = 'turquoise'
pk_no_penalty_color = 'palevioletred'
```

```python
linear_linestyle = 'dashdot'
bilinear_linestyle = 'dotted'
pk_linestyle = 'solid'
```

```python
# Set the base path
weights_path = 'results/kdv/weights_sinpiu'
data_path = 'results/kdv/data_sinpiu'
figures_path = 'results/kdv/figures_sinpiu'
```

```python
import os
```

```python
label_font = 24
ticks_font = 22
legend_font = 24
linewidth = 2.5
title_font = 16
```

# Import prediction results of Linear & Bilinear & Parametric Koopman

## Mass Prediction

```python
mass_dict_error_linear = np.load(os.path.join(data_path,'mass_error_linear.npy'), allow_pickle=True)
mass_mean_linear, mass_std_linear, mass_mean_plus_linear, mass_mean_minus_linear = mass_dict_error_linear[()]['mean'],mass_dict_error_linear[()]['std'], mass_dict_error_linear[()]['mean_plus'], mass_dict_error_linear[()]['mean_minus']

mass_dict_error_bilinear = np.load(os.path.join(data_path,'mass_error_bilinear.npy'), allow_pickle=True)
mass_mean_bilinear, mass_std_bilinear, mass_mean_plus_bilinear, mass_mean_minus_bilinear = mass_dict_error_bilinear[()]['mean'],mass_dict_error_bilinear[()]['std'], mass_dict_error_bilinear[()]['mean_plus'], mass_dict_error_bilinear[()]['mean_minus']

mass_dict_error_pk = np.load(os.path.join(data_path,'mass_error_pk.npy'), allow_pickle=True)
mass_mean_pk, mass_std_pk, mass_mean_plus_pk, mass_mean_minus_pk = mass_dict_error_pk[()]['mean'],mass_dict_error_pk[()]['std'], mass_dict_error_pk[()]['mean_plus'], mass_dict_error_pk[()]['mean_minus']
```

```python
plt.figure(figsize=(6,4))
t_axis = np.arange(0, mass_mean_linear.shape[0], 1)

plt.plot(mass_mean_linear, label='LinearK-DL', color=linear_curve, linewidth=linewidth)
plt.fill_between(t_axis, np.maximum(mass_mean_minus_linear,0), mass_mean_plus_linear, color=linear_shadow, alpha=0.5)


plt.plot(mass_mean_bilinear, label='BilinearK-DL', color=bilinear_curve, linewidth=linewidth)
plt.fill_between(t_axis, np.maximum(mass_mean_minus_bilinear,0), mass_mean_plus_bilinear, color=bilinear_shadow, alpha=0.5)

plt.plot(mass_mean_pk, label='ParamK-DL', color=pk_curve, linewidth=linewidth)
plt.fill_between(t_axis, np.maximum(mass_mean_minus_pk,0), mass_mean_plus_pk, color=pk_shadow, alpha=0.5)

plt.xticks(fontsize=ticks_font)
plt.yticks(fontsize=ticks_font)

plt.xlabel(r'$n$', fontsize=label_font)
plt.ylabel('Error', fontsize=label_font)
# plt.yscale('log')
plt.legend(fontsize=legend_font, loc=(0.03,0.61))

# plt.savefig(os.path.join(figures_path, 'kdv_pred_error.png'), dpi=200, bbox_inches='tight')
# plt.savefig(os.path.join(figures_path, 'kdv_pred_error.pdf'), dpi=200, bbox_inches='tight')
```

# Tracking Problem

## Mass

```python
L = 2 * np.pi
Nx = 128
dx = L / (Nx - 1.0)
```

```python
linear_opt_control_mass = np.loadtxt(os.path.join(data_path,'penalty_linear_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
linear_kdv_opt_mass_soln = np.loadtxt(os.path.join(data_path,'penalty_linear_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
linear_kdv_opt_mass = dx * tf.reduce_sum(linear_kdv_opt_mass_soln, axis=-1)

bilinear_opt_control_mass = np.loadtxt(os.path.join(data_path,'penalty_bilinear_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
bilinear_kdv_opt_mass_soln = np.loadtxt(os.path.join(data_path,'penalty_bilinear_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
bilinear_kdv_opt_mass = dx * tf.reduce_sum(bilinear_kdv_opt_mass_soln, axis=-1)

pk_opt_control_mass = np.loadtxt(os.path.join(data_path,'penalty_pk_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
pk_kdv_opt_mass_soln = np.loadtxt(os.path.join(data_path,'penalty_pk_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
pk_kdv_opt_mass = dx * tf.reduce_sum(pk_kdv_opt_mass_soln, axis=-1)
```

```python
linear_opt_control_mass_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_linear_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
linear_kdv_opt_mass_soln_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_linear_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
linear_kdv_opt_mass_no_penalty = dx * tf.reduce_sum(linear_kdv_opt_mass_soln_no_penalty, axis=-1)

bilinear_opt_control_mass_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_bilinear_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
bilinear_kdv_opt_mass_soln_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_bilinear_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
bilinear_kdv_opt_mass_no_penalty = dx * tf.reduce_sum(bilinear_kdv_opt_mass_soln_no_penalty, axis=-1)

pk_opt_control_mass_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_pk_mass_kdv_opt_control_20s_greedy.csv'), delimiter=',')
pk_kdv_opt_mass_soln_no_penalty = np.loadtxt(os.path.join(data_path,'no_penalty_pk_mass_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
pk_kdv_opt_mass_no_penalty = dx * tf.reduce_sum(pk_kdv_opt_mass_soln_no_penalty, axis=-1)
```

```python
exact_opt_control = np.loadtxt(os.path.join(data_path,'exact_soln_kdv_opt_control_20s_greedy.csv'), delimiter=',')
exact_kdv_opt_soln = np.loadtxt(os.path.join(data_path,'exact_soln_kdv_opt_soln_20s_greedy.csv'), delimiter=',')
exact_kdv_opt_mass = dx * tf.reduce_sum(exact_kdv_opt_soln, axis=-1)
exact_kdv_opt_momentum = dx * tf.reduce_sum(tf.square(exact_kdv_opt_soln), axis=-1)
```

```python
soln_ref = np.loadtxt(os.path.join(data_path,'track_soln_ref'), delimiter=',')
mass_ref = dx * tf.reshape(tf.math.reduce_sum(soln_ref, axis=-1), shape=(-1,1))
```

```python
mass_pk_no_penalty = dx * tf.reshape(tf.math.reduce_sum(pk_kdv_opt_mass_soln_no_penalty, axis=-1), shape=(-1,1))
```

```python
plt.figure(figsize=(10,5))
plt.plot(mass_ref, label='Reference', linestyle='--', color=ref_color, linewidth=linewidth)

plt.plot(linear_kdv_opt_mass, label='LinearK-DL ($\lambda=0.005$)', color=linear_curve, linestyle=linear_linestyle, linewidth=linewidth)
plt.plot(bilinear_kdv_opt_mass, label='BilinearK-DL ($\lambda=0.005$)', color=bilinear_curve,linestyle=bilinear_linestyle, linewidth=linewidth)
plt.plot(pk_kdv_opt_mass, label='ParamK-DL ($\lambda=0.005$)', color=pk_curve,linestyle=pk_linestyle, linewidth=linewidth)

plt.plot(linear_kdv_opt_mass_no_penalty, label='LinearK-DL ($\lambda=0$)',color=linear_no_penalty_color, linestyle=linear_linestyle, linewidth=linewidth)
plt.plot(bilinear_kdv_opt_mass_no_penalty, label='BilinearK-DL ($\lambda=0$)', color=bilinear_no_penalty_color, linestyle=bilinear_linestyle, linewidth=linewidth)
plt.plot(pk_kdv_opt_mass_no_penalty, label='ParamK-DL ($\lambda=0$)', color=pk_no_penalty_color,linestyle='dashed', linewidth=linewidth)

plt.plot(exact_kdv_opt_mass, label='KdV',linestyle='--',color=exact_color, linewidth=linewidth)

plt.xticks(fontsize=ticks_font)
plt.yticks(fontsize=ticks_font)
plt.xlabel(r'$n$', fontsize=label_font)
plt.ylabel('Mass', fontsize=label_font)
# plt.yscale('log')
# plt.legend(fontsize=legend_font, loc=(1.05, 0.1))

plt.savefig(os.path.join(figures_path, 'kdv_mass_track.png'), dpi=200, bbox_inches='tight')
plt.savefig(os.path.join(figures_path, 'kdv_mass_track.pdf'), dpi=200, bbox_inches='tight')
```
