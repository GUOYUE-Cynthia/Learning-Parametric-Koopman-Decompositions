```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot setting
linear_color = 'darkorange'
bilinear_color = 'forestgreen'
pk_color = 'steelblue'
ref_color = 'darkred'
```

# kdv exact solution

```python
kdv_soln_pred = np.loadtxt('results/kdv/data/kdv_soln_pred.csv', delimiter=',')
kdv_mass_pred = tf.reshape(tf.math.reduce_sum(kdv_soln_pred, axis=-1), shape=(-1,1))
kdv_momentum_pred = tf.reshape(tf.math.reduce_sum(tf.square(kdv_soln_pred), axis=-1), shape=(-1,1))
```

# Import Linear & Bilinear & PK prediction results

```python
linear_kdv_mass_pred = np.loadtxt('results/kdv/data/linear_kdv_mass_pred.csv', delimiter=',')
linear_kdv_momentum_pred = np.loadtxt('results/kdv/data/linear_kdv_momentum_pred.csv', delimiter=',')
```

```python
bilinear_kdv_mass_pred = np.loadtxt('results/kdv/data/bilinear_kdv_mass_pred.csv', delimiter=',')
bilinear_kdv_momentum_pred = np.loadtxt('results/kdv/data/bilinear_kdv_momentum_pred.csv', delimiter=',')
```

```python
pk_kdv_mass_pred = np.loadtxt('results/kdv/data/pk_kdv_mass_pred.csv', delimiter=',')
pk_kdv_momentum_pred = np.loadtxt('results/kdv/data/pk_kdv_momentum_pred.csv', delimiter=',')
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

# Plot the data in the first subplot
axs[0].set_title('Mass')
axs[0].plot(kdv_mass_pred, label='Reference', color=ref_color, linestyle='--')
axs[0].plot(linear_kdv_mass_pred, label='Linear', color=linear_color)
axs[0].plot(bilinear_kdv_mass_pred, label='Bilinear', color=bilinear_color)
axs[0].plot(pk_kdv_mass_pred, label='PK', color=pk_color)
axs[0].legend()

# Plot the data in the second subplot
axs[1].set_title('Momentum')
axs[1].plot(kdv_momentum_pred, label='Reference', color=ref_color, linestyle='--')
axs[1].plot(linear_kdv_momentum_pred, label='Linear', color=linear_color)
axs[1].plot(bilinear_kdv_momentum_pred, label='Bilinear', color=bilinear_color)
axs[1].plot(pk_kdv_momentum_pred, label='PK', color=pk_color)
axs[1].legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

# Import Linear & Bilinear & PK tracking results

```python
soln_ref = np.loadtxt('results/kdv/data/track_soln_ref', delimiter=',')
```

```python
linear_opt_control_mass = np.loadtxt('results/kdv/data/linear_mass_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
linear_kdv_opt_mass_soln = np.loadtxt('results/kdv/data/linear_mass_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')

bilinear_opt_control_mass = np.loadtxt('results/kdv/data/bilinear_mass_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
bilinear_kdv_opt_mass_soln = np.loadtxt('results/kdv/data/bilinear_mass_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')

pk_opt_control_mass = np.loadtxt('results/kdv/data/pk_mass_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
pk_kdv_opt_mass_soln = np.loadtxt('results/kdv/data/pk_mass_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')
```

```python
linear_opt_control_momentum = np.loadtxt('results/kdv/data/linear_momentum_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
linear_kdv_opt_momentum_soln = np.loadtxt('results/kdv/data/linear_momentum_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')

bilinear_opt_control_momentum = np.loadtxt('results/kdv/data/bilinear_momentum_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
bilinear_kdv_opt_momentum_soln = np.loadtxt('results/kdv/data/bilinear_momentum_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')

pk_opt_control_momentum = np.loadtxt('results/kdv/data/pk_momentum_kdv_opt_control_30s_tau_2_repeat.csv', delimiter=',')
pk_kdv_opt_momentum_soln = np.loadtxt('results/kdv/data/pk_momentum_kdv_opt_soln_30s_tau_2_repeat.csv', delimiter=',')
```

```python
linear_kdv_opt_mass = tf.reduce_sum(linear_kdv_opt_mass_soln, axis=-1)
linear_kdv_opt_momentum = tf.reduce_sum(tf.square(linear_kdv_opt_momentum_soln), axis=-1)

bilinear_kdv_opt_mass = tf.reduce_sum(bilinear_kdv_opt_mass_soln, axis=-1)
bilinear_kdv_opt_momentum = tf.reduce_sum(tf.square(bilinear_kdv_opt_momentum_soln), axis=-1)

pk_kdv_opt_mass = tf.reduce_sum(pk_kdv_opt_mass_soln, axis=-1)
pk_kdv_opt_momentum = tf.reduce_sum(tf.square(pk_kdv_opt_momentum_soln), axis=-1)
```

```python
mass_ref = tf.reshape(tf.math.reduce_sum(soln_ref, axis=-1), shape=(-1,1))
momentum_ref = tf.reshape(tf.math.reduce_sum(tf.square(soln_ref), axis=-1), shape=(-1,1))
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

# Plot the data in the first subplot
axs[0].set_title('Mass')
axs[0].plot(mass_ref, label='Reference', color=ref_color, linestyle='--')
axs[0].plot(linear_kdv_opt_mass, label='Linear', color=linear_color)
axs[0].plot(bilinear_kdv_opt_mass, label='Bilinear', color=bilinear_color)
axs[0].plot(pk_kdv_opt_mass, label='PK', color=pk_color)
axs[0].legend()

# Plot the data in the second subplot
axs[1].set_title('Momentum')
axs[1].plot(momentum_ref, label='Reference', color=ref_color, linestyle='--')
axs[1].plot(linear_kdv_opt_momentum, label='Linear', color=linear_color)
axs[1].plot(bilinear_kdv_opt_momentum, label='Bilinear', color=bilinear_color)
axs[1].plot(pk_kdv_opt_momentum, label='PK', color=pk_color)
axs[1].legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
