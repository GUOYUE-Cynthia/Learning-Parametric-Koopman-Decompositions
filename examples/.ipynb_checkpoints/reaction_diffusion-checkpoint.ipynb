```python
import numpy as np
from scipy.integrate import solve_ivp

# Define the reaction-diffusion equation with boundary conditions
def reaction_diffusion_equation(t,u, temp):
    D = 1  # Diffusion coefficient
    
    lambda1 = 1
    lambda2 = 2
    
    a = np.exp(-lambda1/temp) # Reaction coefficient
    b = np.exp(-lambda2/temp) # Reaction coefficient
    
#     a = temp # Reaction coefficient
#     b = temp # Reaction coefficient

  
    # Calculate the second derivative using finite differences
    d2u_dx2 = np.diff(u, 2) / dx**2

    # Apply boundary conditions
    d2u_dx2 = np.concatenate(([0], d2u_dx2, [0]))

    # Calculate the reaction term
    reaction = a * u + b * u**3 

    # Combine the diffusion and reaction terms
    du_dt = D * d2u_dx2 - reaction +1

    return du_dt
```

```python
# Define the spatial domain
L = 1.0  # Length of the domain
Nx = 50  # Number of spatial points
dx = L / (Nx - 1)  # Spatial step size
T = 1
Nt = 50
dt = T/Nt
```

```python
w = 5
k = 0.01
t = np.linspace(0, T, Nt-1)
temp_list = 10 * np.exp(- k * t) * np.abs(np.sin(np.pi * w * t)) + 0.01
```

```python
soln = [u0]
for temp in temp_list:
    # Solve the equation using the scipy integrator
    curr_soln = solve_ivp(fun=reaction_diffusion_equation,
                         t_span=t_span, 
                         y0=soln[-1], 
                         args=(temp,))
    soln.append(curr_soln.y.T[-1])

soln = np.asarray(soln)
```

# Koopman

```python
target_dim = Nx # dim of mean = 1
param_dim = 1
```

```python
n_psi_train = 10
n_psi = 1 + 1 + n_psi_train
```

```python
from koopmanlib.dictionary import PsiNN_obs
from koopmanlib.param_solver import KoopmanParametricDLSolver
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
dic_pk = PsiNN_obs(layer_sizes=[128,128], n_psi_train=n_psi_train)
```

```python
solver_pk = KoopmanParametricDLSolver(target_dim=target_dim, 
                                      param_dim=param_dim, 
                                      n_psi=n_psi,
                                      dic=dic_pk)
```

```python
model_pk, model_K_u_pred_pk = solver_pk.generate_model(layer_sizes=[16,16])
```

```python
model_pk.compile(optimizer=Adam(0.001),
             loss='mse')
```

```python
lr_callbacks = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss',
                                                    factor=0.1,
                                                    patience=50,
                                                    verbose=0,
                                                    mode='auto',
                                                    min_delta=0.0001,
                                                    cooldown=0,
                                                    min_lr=1e-12)
```

```python
zeros_data_z_next_train = tf.zeros_like(dic_pk(data_y))
```

```python
history = model_pk.fit(x=[data_x, data_y, data_temp], 
                    y=zeros_data_z_next_train, 
                    epochs=300, 
                    batch_size=200,
                    callbacks=lr_callbacks,
                    verbose=1)
```

```python
B = dic_pk.generate_B(mean_ref)
```

```python
param_init = np.random.uniform(size=temp_list.shape)
```

```python
def mpc_loss_pk(param, tau, ref_list, u0, current_time):
    param = tf.reshape(param, shape=(param.shape[0], 1, param_dim))
    loss_list = []
    u0 = u0.reshape(1,-1)
    psi_x = dic_pk(u0)
    
    for i in range(tau):
        psi_x = model_K_u_pred_pk([param[i], psi_x])
        y_next = psi_x@B
        loss_curr = tf.square(tf.norm(ref_list[current_time+i] - y_next))
        loss_list.append(loss_curr)
    
    ref_loss= tf.reduce_sum(loss_list)
    param_loss = 0.001 * tf.square(tf.norm(param))
    
    loss = ref_loss
    return loss   
```

```python
current_time = 0
tau = temp_list.shape[0]
```

```python
param_init = np.random.uniform(size=temp_list.shape)
results = minimize(mpc_loss_pk, 
              x0=param_init,
              args = (tau, mean_ref, u0, current_time))
```
