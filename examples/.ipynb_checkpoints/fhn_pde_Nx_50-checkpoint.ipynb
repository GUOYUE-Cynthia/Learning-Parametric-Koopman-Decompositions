# Generate data

```python
Nx = 50
n_init = 100
traj_len = 500
```

```python
x = np.linspace(-10,10,Nx)
t = np.arange(0, traj_len, 1)
```

```python
target_dim = Nx*2
param_dim = 1
```

```python
n_psi_train = 10
```

```python
n_psi = 1 + target_dim + n_psi_train
```

```python
from koopmanlib.target import FitzHughNagumoTarget
```

```python
fhn_pde = FitzHughNagumoTarget(
            n_init=n_init,
            traj_len=traj_len,
            x=x,
            dt=1e-5,
            t_step=1e-3,
            dim=target_dim,
            param_dim=param_dim,
            k=2,
            seed_z=12,
            seed_param=12)
```

```python
data_z_curr, data_u, data_z_next = fhn_pde.generate_data()

data_z_curr.shape

data_z_curr_reshape = data_z_curr.reshape(-1,n_init,target_dim)
data_z_curr_reshape_trans = np.transpose(data_z_curr_reshape, (1,0,2))

data_z_next_reshape = data_z_next.reshape(-1,n_init,target_dim)
data_z_next_reshape_trans = np.transpose(data_z_next_reshape, (1,0,2))

sample_idx = 1
```

```python
x_axis = x
y_axis = np.arange(0, t.shape[0]-1, 1)

# Random z(x, t0)
X, Y = np.meshgrid(x_axis, y_axis)
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5), constrained_layout=True)
c1 = ax1.pcolor(X,Y,data_z_curr_reshape_trans[sample_idx][:,:Nx])
fig.colorbar(c1, ax=ax1)
ax1.set_xlabel('space')
ax1.set_ylabel('time')
ax1.set_title('$v$')

c2 = ax2.pcolor(X,Y,data_z_curr_reshape_trans[sample_idx][:,Nx:])
fig.colorbar(c2, ax=ax2)
ax2.set_xlabel('space')
ax2.set_ylabel('time')
ax2.set_title('$w$')


```

# Build model

```python
from koopmanlib.dictionary import PsiNN
from koopmanlib.param_solver import KoopmanParametricDLSolver
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
dic_pk = PsiNN(layer_sizes=[16,16], n_psi_train=n_psi_train)
```

```python
solver_pk = KoopmanParametricDLSolver(target_dim=target_dim, 
                                      param_dim=param_dim, 
                                      n_psi=n_psi,
                                      dic=dic_pk)
```

```python
model_pk, model_K_u_pred_pk = solver_pk.generate_model(layer_sizes=[16])
```

```python
def my_loss(y_true, y_pred):
    w_true = y_true[:,Nx+1:2*Nx+1]
    w_pred = y_pred[:,Nx+1:2*Nx+1]
    squared_diff = tf.square(y_true - y_pred)
    additional_diff = tf.square(w_true - w_pred)
    mse = tf.reduce_mean(squared_diff, axis=-1)
    additional_mse = tf.reduce_mean(additional_diff, axis=-1)
    return mse + 1e3*additional_mse
```

```python
# model_pk.summary()
```

```python
model_pk.compile(optimizer=Adam(0.001),
             loss='mse')
```

```python
lr_callbacks = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss',
                                                    factor=0.1,
                                                    patience=50,
                                                    verbose=0,
                                                    mode='auto',
                                                    min_delta=0.0001,
                                                    cooldown=0,
                                                    min_lr=1e-12)
```

# Prediction

```python
n_init_test = 1
traj_len_test = 100
```

```python
def compute_diff_ratio_one_traj(data_true_list, data_pred_list):
    diff = data_true_list - data_pred_list
    diff_norm = np.linalg.norm(diff, axis=-1)
    data_true_norm = np.linalg.norm(data_true_list, axis=-1)

    sum_diff_norm_list = []
    sum_data_true_norm_list = []
    for i in range(traj_len_test):
        sum_diff_norm = np.sum(diff_norm[:i+1])
        sum_data_true_norm = np.sum(data_true_norm[:i+1])
        sum_diff_norm_list.append(sum_diff_norm)
        sum_data_true_norm_list.append(sum_data_true_norm)

    sum_diff_norm_list = np.asarray(sum_diff_norm_list)
    sum_data_true_norm_list = np.asarray(sum_data_true_norm_list)

    ratio = sum_diff_norm_list / sum_data_true_norm_list
    return ratio
```

```python
fhn_pde_test = FitzHughNagumoTarget(
            n_init=n_init_test,
            traj_len=traj_len_test,
            x=x,
            dt=1e-5,
            t_step=1e-3,
            dim=target_dim,
            param_dim=param_dim,
            k=2,
            seed_z=2,
            seed_param=3)

data_z_curr_test, data_u_test, data_z_next_test = fhn_pde_test.generate_data()

data_z_init = tf.reshape(data_z_curr_test[0], shape=(1,-1))
data_true = data_z_curr_test
data_pred_list_pk= solver_pk.compute_data_list(dic_pk, 
                                 model_K_u_pred_pk, 
                                 traj_len_test,
                                 data_z_init,
                                 data_u_test)
```

```python
data_pred_pk = data_pred_list_pk[:-1,:]
data_pred_pk.shape
```

```python
x_axis = x
y_axis = np.arange(0, data_true.shape[0], 1)
```

```python
X, Y = np.meshgrid(x_axis, y_axis)
fig, axs = plt.subplots(2,2, figsize=(16,12))
ax1, ax2, ax3, ax4 = axs[0,0], axs[0,1], axs[1,0], axs[1,1]
c1 = ax1.pcolormesh(X,Y,data_true[:,:Nx])
fig.colorbar(c1, ax=ax1)
ax1.set_xlabel('space')
ax1.set_ylabel('time')
# ax1.set_ylim(bottom=0, top=20)
ax1.set_title('$v$ (true)')

c2 = ax2.pcolormesh(X,Y,data_pred_pk[:,:Nx])
fig.colorbar(c2, ax=ax2)
ax2.set_xlabel('space')
ax2.set_ylabel('time')
ax2.set_title('$v$ (prediction)')

c3 = ax3.pcolormesh(X,Y,data_true[:,Nx:])
fig.colorbar(c3, ax=ax3)
ax3.set_xlabel('space')
ax3.set_ylabel('time')
ax3.set_title('$w$ (true)')


c4 = ax4.pcolormesh(X,Y,data_pred_pk[:,Nx:])
fig.colorbar(c4, ax=ax4)
ax4.set_xlabel('space')
ax4.set_ylabel('time')
ax4.set_title('$w$ (prediction)')
```

```python
X, Y = np.meshgrid(x_axis, y_axis)
fig, axs = plt.subplots(1,2, figsize=(16,6))
ax1, ax2 = axs[0], axs[1]
c1 = ax1.pcolormesh(X,Y,data_true[:,:Nx]-data_pred_pk[:,:Nx])
fig.colorbar(c1, ax=ax1)
ax1.set_xlabel('space')
ax1.set_ylabel('time')
# ax1.set_ylim(bottom=0, top=20)
ax1.set_title('Difference $v$')

c2 = ax2.pcolormesh(X,Y,data_true[:,Nx:]-data_pred_pk[:,Nx:])
fig.colorbar(c2, ax=ax2)
ax2.set_xlabel('space')
ax2.set_ylabel('time')
ax2.set_title('Difference $w$')
```

# Average error

```python
np.random.seed(100)
n_test = 20
seed_z_list = np.random.randint(1,300, size=(n_test,))
seed_param_list = np.random.randint(1,300, size=(n_test,))
```

```python
n_traj_test = 1
```

```python
ratio_pk_list = []
```

```python
traj_pred_len = traj_len_test-1
```

```python
x_axis = x
y_axis = np.arange(0, traj_pred_len, 1)
```

```python
for seed_z, seed_param in zip(seed_z_list, seed_param_list):
    fhn_pde_test = FitzHughNagumoTarget(
            n_init=n_traj_test,
            traj_len=traj_len_test,
            x=x,
            dt=1e-5,
            t_step=1e-3,
            dim=target_dim,
            param_dim=param_dim,
            k=2,
            seed_z=seed_z,
            seed_param=seed_param)

    data_z_curr_test, data_u_test, data_z_next_test = fhn_pde_test.generate_data()

    data_z_init = tf.reshape(data_z_curr_test[0], shape=(1,-1))
    data_true = data_z_curr_test
    data_pred_list_pk= solver_pk.compute_data_list(dic_pk, 
                                     model_K_u_pred_pk, 
                                     traj_len_test,
                                 data_z_init,
                                 data_u_test)
    
    data_pred_pk = data_pred_list_pk[:-1,:]
    

    
    X, Y = np.meshgrid(x_axis, y_axis)
    fig, axs = plt.subplots(1,2, figsize=(16,6))
    ax1, ax2 = axs[0], axs[1]
    c1 = ax1.pcolormesh(X,Y,data_true[:,:Nx]-data_pred_pk[:,:Nx])
    fig.colorbar(c1, ax=ax1)
    ax1.set_xlabel('space')
    ax1.set_ylabel('time')
    # ax1.set_ylim(bottom=0, top=20)
    ax1.set_title('Difference $v$')

    c2 = ax2.pcolormesh(X,Y,data_true[:,Nx:]-data_pred_pk[:,Nx:])
    fig.colorbar(c2, ax=ax2)
    ax2.set_xlabel('space')
    ax2.set_ylabel('time')
    ax2.set_title('Difference $w$')
    fig.show()
    
    
    ratio_pk = compute_diff_ratio_one_traj(data_true, 
                                    data_pred_pk)
    
    
    
    ratio_pk_list.append(ratio_pk)
    
```

```python
ratio_pk_list_max = np.max(ratio_pk_list, axis=0)
ratio_pk_list_min = np.min(ratio_pk_list, axis=0)
```

```python
ratio_pk_list_std = np.std(ratio_pk_list, axis=0)
```

```python
ratio_pk_list_mean = np.mean(ratio_pk_list, axis=0)
```

```python
t_axis = np.arange(0, traj_pred_len+1, 1)
```

```python
plt.plot(ratio_pk_list_mean, label='P-Koopman mean', color='forestgreen')
plt.fill_between(t_axis, ratio_pk_list_min, ratio_pk_list_max, color='darkseagreen')
# plt.plot(ratio_pk_list_max, label='P-Koopman max')
# plt.plot(ratio_pk_list_min, label='P-Koopman min')
# plt.plot(ratio_pk_list_mean+ratio_pk_list_std, label='P-Koopman + std')
# plt.plot(ratio_pk_list_mean-ratio_pk_list_std, label='P-Koopman - std')
plt.legend()
```
