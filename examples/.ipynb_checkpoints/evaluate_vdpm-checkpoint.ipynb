```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
mean_dict_mu_0 = np.load('results/vdpm/ratio_list_dict_mu_0_mean.npy', allow_pickle=True)
mean_dict_mu_1 = np.load('results/vdpm/ratio_list_dict_mu_1_mean.npy', allow_pickle=True)
mean_dict_mu_2 = np.load('results/vdpm/ratio_list_dict_mu_2_mean.npy', allow_pickle=True)
mean_dict_mu_3 = np.load('results/vdpm/ratio_list_dict_mu_3_mean.npy', allow_pickle=True)
mean_dict_mu_4 = np.load('results/vdpm/ratio_list_dict_mu_4_mean.npy', allow_pickle=True)
```

```python
std_dict_mu_0 = np.load('results/vdpm/ratio_list_dict_mu_0_std.npy', allow_pickle=True)
std_dict_mu_1 = np.load('results/vdpm/ratio_list_dict_mu_1_std.npy', allow_pickle=True)
std_dict_mu_2 = np.load('results/vdpm/ratio_list_dict_mu_2_std.npy', allow_pickle=True)
std_dict_mu_3 = np.load('results/vdpm/ratio_list_dict_mu_3_std.npy', allow_pickle=True)
std_dict_mu_4 = np.load('results/vdpm/ratio_list_dict_mu_4_std.npy', allow_pickle=True)
```

```python
def extact_data(ratio_dict, ratio_name):
    ratio_pk = ratio_dict[()][ratio_name[0]].reshape((1,-1))
    ratio_linear = ratio_dict[()][ratio_name[1]].reshape((1,-1))
    ratio_bilinear = ratio_dict[()][ratio_name[2]].reshape((1,-1))
    return ratio_pk, ratio_linear, ratio_bilinear
```

```python
mean_dict_list = [mean_dict_mu_0, mean_dict_mu_1, mean_dict_mu_2, mean_dict_mu_3, mean_dict_mu_4]
```

```python
std_dict_list = [std_dict_mu_0, std_dict_mu_1, std_dict_mu_2, std_dict_mu_3, std_dict_mu_4]
```

```python
ratio_pk_list = []
ratio_linear_list = []
ratio_bilinear_list = []
```

```python
ratio_name_list = ['pk', 'linear', 'bilinear']
```

```python
def concat_ratio_list(ratio_list):
    ratio_concat = np.concatenate(ratio_list, axis=0)
    ratio_concat = ratio_concat.T
    return ratio_concat
```

```python
def concat_data(dict_list):
    pk_list=[]
    linear_list = []
    bilinear_list = []

    for i in range(5):
        pk, linear, bilinear = extact_data(dict_list[i], ratio_name=ratio_name_list)
        pk_list.append(pk)
        linear_list.append(linear)
        bilinear_list.append(bilinear)
        
    pk_concat = concat_ratio_list(pk_list)
    linear_concat = concat_ratio_list(linear_list)
    bilinear_concat = concat_ratio_list(bilinear_list)
    
    return pk_concat, linear_concat, bilinear_concat
```

```python
mean_pk, mean_linear, mean_bilinear = concat_data(mean_dict_list)
std_pk, std_linear, std_bilinear = concat_data(std_dict_list)
```

```python
std_pk.shape
```

```python
mean_linear_mean = np.mean(mean_linear, axis=0)
std_linear_mean = np.mean(std_linear, axis=0)

mean_bilinear_mean = np.mean(mean_bilinear, axis=0)
std_bilinear_mean = np.mean(std_bilinear, axis=0)

mean_pk_mean = np.mean(mean_pk, axis=0)
std_pk_mean = np.mean(std_pk, axis=0)
```

```python
mean_plus_linear = mean_linear_mean + std_linear_mean
mean_minus_linear = mean_linear_mean - std_linear_mean

mean_plus_bilinear = mean_bilinear_mean + std_bilinear_mean
mean_minus_bilinear = mean_bilinear_mean - std_bilinear_mean

mean_plus_pk = mean_pk_mean + std_pk_mean
mean_minus_pk = mean_pk_mean - std_pk_mean
```

```python
mu = [0,1,2,3,4]
mu_axis = list(range(len(mu)))
```

```python
font = {'size': 18}
```

```python
plt.figure(figsize=(15,4))
plt.plot(mu, mean_pk_mean, label='P-Koopman', marker='p', color='darkred', linestyle='dashed', linewidth=3)
# plt.fill_between(mu_axis, mean_minus_pk, mean_plus_pk, color='lightsteelblue', alpha=0.5)
plt.plot(mu, mean_linear_mean, label='Linear', marker='*', color='tomato', linewidth=3)
# plt.fill_between(mu_axis, mean_minus_linear, mean_plus_linear, color='navajowhite', alpha=0.5)
plt.plot(mu, mean_bilinear_mean, label='Bilinear', marker='s', color='rebeccapurple', linestyle='dashdot', linewidth=3)
# plt.fill_between(mu_axis, mean_minus_bilinear, mean_plus_bilinear, color='darkseagreen', alpha=0.5)

plt.xlabel(r'$\mu$', loc='right', fontsize=24)
plt.ylabel('Error', fontsize=24)
plt.xticks(mu_axis, mu, fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=24)

# plt.savefig('results/vdpm/figures/vdpm_comparison_mu.png', dpi=200, bbox_inches='tight')
# plt.savefig('results/vdpm/figures/vdpm_comparison_mu.pdf', dpi=200, bbox_inches='tight')
```
