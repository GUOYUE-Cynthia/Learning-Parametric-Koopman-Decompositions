```python
class EinsumLayer(tf.keras.layers.Layer):
    """
    Layer wrapping a single tf.einsum operation.

    Usage:
    psi_y = EinsumLayer("ij,ijk->ik")((psi_x, K_u))
    """

    def __init__(self, equation: str):
        super().__init__()
        self.equation = equation

    def call(self, psi_x, K_u):
        return tf.einsum(self.equation, psi_x, K_u)

    def get_config(self):
        return {"equation": self.equation}
```

```python
def generate_model():
    layer_sizes=[128, 256, 128]
    hidden_layers = [Dense(s, activation='tanh') for s in layer_sizes]
    # hidden_layers = [Dense(s) for s in layer_sizes]
    output_layer = Dense(n_phi)
    
    inputs_u = Input((param_dim,))
    hidden_u = inputs_u
    for layer in hidden_layers:
        hidden_u = layer(hidden_u)   
    phi_u = output_layer(hidden_u)
    K_u_entry = Dense(n_psi**2, use_bias=True)(phi_u)
    K_u = tf.reshape(K_u_entry, shape=(-1, n_psi, n_psi))
    model_K_u = Model(inputs=inputs_u, outputs=K_u, name='K_u')
    
    inputs_u = Input((param_dim,))
    inputs_psi_x = Input((n_psi,))
    K_u = model_K_u(inputs_u)
    psi_y = EinsumLayer('ij,ijk->ik')(inputs_psi_x, K_u)
    model_K_u_pred = Model(inputs=[inputs_u, inputs_psi_x], outputs=psi_y, name='K_u_pred')
    
    dic_func = PsiNN(layer_sizes=[100,100,100], n_psi_train=n_psi_train)
    
    # Build the model related to Psi
    inputs_x = Input((target_dim,))
    inputs_y = Input((target_dim,))
    inputs_u = Input((param_dim,))

    psi_x = dic_func(inputs_x)
    psi_y = dic_func(inputs_y)

    psi_next = model_K_u_pred([inputs_u, psi_x])

    outputs = psi_next - psi_y
    model = Model(inputs=[inputs_x, inputs_y, inputs_u], outputs=outputs)
    return model, model_K_u, model_K_u_pred, dic_func
```

```python
model_complex, model_K_u_complex, model_K_u_pred_complex, dic_func_complex = generate_model()
```

```python
from scipy import signal
def generate_control(traj_len):
    # traj_len is the number of iteration times
    # the signal changes every 30 steps (0.01s * 30 = 0.3s)
    interval = 30
    t = np.linspace(0, 1, traj_len, endpoint=False)
    scale = (traj_len / 2) / interval
    # If we set scale as traj_len / 2, then the control only changes once at the middle of the whole time period.
    # We need to rescale t period so that the control changes every “interval” number of steps.
    control = - signal.square(2 * np.pi * scale * t, duty=0.5)
    # plt.plot(0.01 * traj_len * t, control)
    return control
```

```python
# This is multi-step optimization, not MPC. We use tau = 1 as the case for one-step optimization.
def mpc_loss(param, tau, model_K_u_pred, psi_x_true_list, psi_x_pred_list, current_time):
    param = tf.reshape(param, shape=(param.shape[0], 1, u_dim))
    loss_list = []
    psi_next_list = [tf.reshape(psi_x_true_list[current_time], shape=(1,-1))]

    for i in range(tau):
        psi_current = psi_next_list[-1]
        K_psi_next = model_K_u_pred([param[i], psi_current])
        psi_next_list.append(K_psi_next)
        loss_current = tf.norm(K_psi_next-psi_x_true_list[current_time+i+1])
        loss_list.append(loss_current)
    loss= tf.reduce_sum(loss_list)
    return loss
```

```python
tau = 1
model_K_u_pred = model_K_u_pred_complex
psi_x_true_list = dic_func_complex(data_true)
psi_x_pred_list = [dic_func_complex(tf.reshape(data_true[0], shape=(-1, x_dim)))]
current_time = 0
param_init = np.random.uniform(size=(tau*u_dim,))
```

```python
def generate_cons_lhs(i):
    def cons_func_lhs(u):
        return u[i] + 1
    return cons_func_lhs

def generate_cons_rhs(i):
    def cons_func_rhs(u):
        return - u[i] + 1
    return cons_func_rhs
```

```python
cons_list = []
for i in range(tau*u_dim):
    cons_lhs = generate_cons_lhs(i)
    cons_rhs = generate_cons_rhs(i)
    cons_dic_lhs = {'type': 'ineq', 'fun': cons_lhs}
    cons_dic_rhs = {'type': 'ineq', 'fun': cons_rhs}
    cons_list.append(cons_dic_lhs)
    cons_list.append(cons_dic_rhs)
```

```python
results = minimize(mpc_loss, 
                  x0=param_init,
                  args = (tau, model_K_u_pred, psi_x_true_list, psi_x_pred_list, current_time),
                  constraints = cons_list,
                  tol=1e-6)
#                   method='BFGS')
```
