```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import odeint
from scipy.fftpack import diff as psdiff
```

```python
def kdv_exact(x, seed):
    """Profile of the exact solution to the KdV for a single soliton on the real line."""
    
    # Set the seed
    np.random.seed(seed)
    
    b = np.random.uniform(size=(3,))
    b0 = b[0] / np.sum(b)
    b1 = b[1] / np.sum(b)
    b2 = b[2] / np.sum(b)
    
    IC0 = np.exp(-(x-np.pi/2)**2)
    IC1 = -np.sin((x/2)**2)
    IC2 = np.exp(-(x+np.pi/2)**2)
    
    IC = b0*IC0 + b1*IC1 + b2*IC2
    return IC

def kdv(y, t, L, param, v_list):
    """Differential equations for the KdV equation, discretized in x."""
    # Compute the x derivatives using the pseudo-spectral method.
    yx = psdiff(y, period=L)
    yxxx = psdiff(y, period=L, order=3)

    # Compute du/dt.    
    dydt = -y*yx - yxxx
    
    param = param.reshape(1,-1)
    
    param_spatial = param @ v_list
    
    param_spatial = param_spatial.reshape(-1, )
    
    rhs = dydt + param_spatial

    return rhs

def kdv_solution(y0, t, L, param, v_list):
    """Use odeint to solve the KdV equation on a periodic domain.
    
    `y0` is initial condition, `t` is the array of time values at which
    the solution is to be computed, and `L` is the length of the periodic
    domain."""

    sol = odeint(kdv, y0, t, args=(L, param, v_list, ), mxstep=5000)
    return sol
```

```python
# if __name__ == "__main__":
# Set the size of the domain, and create the discretized grid.
L = 2 * np.pi
Nx = 128
dx = L / (Nx - 1.0)
x = np.linspace(-np.pi, np.pi, Nx)

# Set the initial conditions.
y0 = kdv_exact(x, seed=123) 
```

```python
# inner loop
# Set the time sample grid.
dt = 0.4 / (Nx**2)
Nt = int(0.01 / dt)
dt = 0.01 / Nt
T = Nt*dt
t = np.linspace(0, T, Nt)
```

```python
def v_func(x, c):
    return np.exp(-25 * (x - c)**2)

c1, c2, c3 = -np.pi/2, 0, np.pi/2
v1 = v_func(x, c1).reshape(1,-1)
v2 = v_func(x, c2).reshape(1,-1)
v3 = v_func(x, c3).reshape(1,-1)

v_list = np.concatenate([v1,v2,v3], axis=0)
```

```python
umax = 1
umin = -umax
```

```python
traj_length = 50
```

```python
param_list = np.random.uniform(low=0, high=1, size=(traj_length,3)) * (umax - umin) + umin
```

```python
# outer loop
soln_list = [y0]
for i in range(traj_length):
    param = param_list[i]
    soln = kdv_solution(y0, t, L, param, v_list)
    y0 = soln[-1]
    soln_list.append(y0)
```

```python
def mpc_loss_pk(param, tau, ref_list, y0, current_time):
    param = tf.reshape(param, shape=(int(param.shape[0]/param_dim), 1, param_dim))
    loss_list = []
    y0 = y0.reshape(1,-1)
    psi_x = dic_pk(y0)
    
    for i in range(tau):
        psi_x = model_K_u_pred_pk([param[i], psi_x])
        y_next = psi_x@B
#         print('shape1', ref_list[current_time+i].shape)
#         print('shape2', y_next.shape)
        loss_curr = tf.square(tf.norm(ref_list[current_time+i] - y_next))
        loss_list.append(loss_curr)
    
    ref_loss= tf.reduce_sum(loss_list)
    param_loss = 0.001 * tf.square(tf.norm(param))
    
    loss = ref_loss
    return loss   
```

# Substitute control in Koopman model

```python
def compute_koopman(y0, param_list):
    param = tf.reshape(param_list, shape=(param_list.shape[0], 1, param_dim))
    y0 = y0.reshape(1,-1)
    y_list = [y0]
    psi_x = dic_pk(y0)
    
    tau = param_list.shape[0]

    for i in range(tau):
        psi_x = model_K_u_pred_pk([param[i], psi_x])
        y_next = psi_x@B
        y_list.append(y_next)
        
    return np.squeeze(np.asarray(y_list))
```

```python
y0_test = np.zeros(shape=(Nx, ))
pk_test_ori_control = compute_koopman(y0=y0_test, param_list=param_list_ref)
```
